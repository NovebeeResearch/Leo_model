```python
#@title Input prompts and generate the results
import clipit
import gc
gc.collect()
#parameters setting and prompts input
#----------------------------------------------------------------------------------------------
# To reset settings to default
clipit.reset_settings()
#@markdown if error "module 'clipit' has no attribute 'reset_settings'" occur, 
#@markdown please restart everything, it is a issue would happen sometimes. the writer said.
#@markdown https://github.com/mfrashad/text2art/issues/3
# You can use "|" to separate multiple prompts
prompts = "city skyline at dawn" #@param {type:"string"}
# You can trade off speed for quality: draft, normal, better, best
quality = "normal" #@param ["draft", "normal", "better", "best"]
# Aspect ratio: widescreen, square
aspect = "widescreen" #@param ["widescreen", "square"]
#----------------------------------------------------------------------------------------------
# Add settings
clipit.add_settings(prompts=prompts, quality=quality, aspect=aspect)
# Apply these settings and run
settings = clipit.apply_settings()
clipit.do_init(settings)
clipit.do_run(settings)
#@markdown For saving the image, you can simply right click the image and save it to your computer
```
